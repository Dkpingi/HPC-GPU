```python
import matplotlib.pyplot as plt
import csv
from matplotlib.pyplot import figure
figure(num=None, figsize=(12, 9), dpi=80, facecolor='w', edgecolor='k')


# 5.3 Task 1
BlockSize = []
MatrixCalcDur = []

with open('5_3_01_benchmark_1024.txt','r') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader, None) 
    for row in reader:
        BlockSize.append(int(row[1]))
        MatrixCalcDur.append(float(row[5]))
        
plt.plot(BlockSize, MatrixCalcDur, label='Matrix Calculation Time')

plt.xlabel('Block Size')
plt.ylabel('Time [ms]')
plt.title('Exc 5.3 Task 1 with Matrix Size = 1024')
plt.legend()
#plt.savefig("5_3_01_1024.png")
plt.show()
```
