```python

X = pd.read_csv("matMulsize.txt", delimiter = " ", index_col = False)
X = pd.DataFrame(X)
X.rename({"Unnamed: 7":"a"}, axis="columns", inplace=True)
X.drop(["a"], axis=1, inplace=True)
X["TotalTime"] = X["H2DTime"] + X ["D2HTime"] + X["CalcTime"]
Shared = X.loc[X["BoolShMemory"] == 1]
Global = X.loc[X["BoolShMemory"] == 0]
plt.title("MMult Kernel with shared memory with 27 threads")
plt.plot(Shared["MatrixWidth"],Shared["H2DTime"], label = "H2D")
plt.plot(Shared["MatrixWidth"],Shared["D2HTime"], label = "D2H")
plt.plot(Shared["MatrixWidth"],Shared["CalcTime"], label = "Calc")
plt.plot(Shared["MatrixWidth"],Shared["TotalTime"], label = "Total")
plt.xlabel("MatrixWidth")
plt.ylabel("RunTime (ms)")
plt.semilogy()
plt.legend()
plt.savefig("plots/matMuldatash.png")
plt.show()
plt.title("MMult Kernel with global memory with 9 threads")
plt.plot(Global["MatrixWidth"],Global["H2DTime"], label = "H2D")
plt.plot(Global["MatrixWidth"],Global["D2HTime"], label = "D2H")
plt.plot(Global["MatrixWidth"],Global["CalcTime"], label = "Calc")
plt.plot(Global["MatrixWidth"],Global["TotalTime"], label = "Total")
plt.xlabel("MatrixWidth")
plt.ylabel("RunTime (ms)")
plt.semilogy()
plt.legend()
plt.savefig("plots/matMuldatagl.png")
plt.show() 

print("The CPU-Run with 8192 MatrixWidth took 949775 ms \
      compared to the fastest speed with global memory of 60190ms \
      and with shared memory of 13542ms.")
print("Speedups:" + str(949775/60190) + "," +str(949775/13542))
print(X)
```
