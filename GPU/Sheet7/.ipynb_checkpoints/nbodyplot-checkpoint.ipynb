```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv("nBody.txt", delimiter = " ")
dfopt = pd.read_csv("nBody_opt.txt", delimiter = " ")
X = df["numElements"]
Y = df["bodyBodyInt[G]"]
Xopt = dfopt["numElements"]
Yopt = dfopt["bodyBodyInt[G]"]

plt.plot(X,Y,label = "unoptimized")
plt.plot(Xopt,Yopt, label = "optimized")
plt.title("nBodyOptimization")

plt.xlabel("numElements")
plt.ylabel("bodyBodyInt[G]")
plt.legend()

plt.savefig("plots/nBodyOptimization.png")
plt.show()

```
