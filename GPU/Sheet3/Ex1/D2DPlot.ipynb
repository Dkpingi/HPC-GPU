```python
import numpy as np
import matplotlib.pyplot as plt

MemSpeedD2D = np.loadtxt("MemorySpeedD2D.txt", delimiter = " ")

plt.title("D2D Bandwidth for 1e8 data copies")
plt.plot(MemSpeedD2D[:,0],MemSpeedD2D[:,1], label = "D2D")
plt.legend()
plt.xlabel("copy size (Byte)")
plt.ylabel("bandwidth (Byte/s)")
plt.semilogy()
plt.savefig("plots/D2D.png")
plt.show()
```
