```python
import numpy as np
import matplotlib.pyplot as plt

Onenode = np.loadtxt("Ex1resonenode.txt",delimiter = " ")
Twonodes = np.loadtxt("Ex1restwonodes.txt",delimiter = " ")
Onenode *= 1e6
Twonodes *= 1e6
plt.title("Ping-Pong time per message for m messages")
plt.semilogy()
plt.semilogx()
plt.ylabel("time (us)")
plt.xlabel("message size (byte)")
plt.plot(Onenode[:,0],Onenode[:,1],label = "One Node (m = 1000000)")
plt.plot(Twonodes[:,0],Twonodes[:,1],label = "Two Nodes (m = 10000)")
plt.legend()
plt.show()
```
