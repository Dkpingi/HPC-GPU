```python
import numpy as np
import matplotlib.pyplot as plt

g2s = np.loadtxt("global2shared.txt", delimiter = " ")

g2sth= g2s[:,1:]
#print(g2sth)
g2sth = g2sth[g2sth[:,0].argsort()]
#print(g2sth)
plt.plot(g2sth[:,0],g2sth[:,2],label = "Threads")
plt.show()
plt.plot(g2s[:,1],g2s[:,2],label = "datasize")
plt.show()
X = g2sth[:,0]
Y = g2sth[:,1]
Z = g2sth[:,2]
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(X,Y,Z)
ax.set_xlabel("nthreads (1 Block)")
ax.set_ylabel("data (bytes)")
ax.set_zlabel("bandwidth (GB/s)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
g2s = np.loadtxt("global2shared.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('blockdim').mean().reset_index()
print(df2)
plt.plot(df2['blockdim'],df2['bandwidth'],label = "g2s")

g2s = np.loadtxt("shared2global.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('blockdim').mean().reset_index()
print(df2)
plt.plot(df2['blockdim'],df2['bandwidth'],label = "s2g")
plt.legend()
plt.xlabel("blockdim")
plt.ylabel("bandwidth (GB/s)")
plt.show()

g2s = np.loadtxt("global2sharedblocks.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
print(df)
plt.plot(df['griddim'],df['bandwidth'],label = "g2s")

g2s = np.loadtxt("shared2globalblocks.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
print(df)
plt.plot(df['griddim'],df['bandwidth'],label = "s2g")
plt.legend()
plt.xlabel("griddim")
plt.ylabel("bandwidth (GB/s)")
plt.show()
```
