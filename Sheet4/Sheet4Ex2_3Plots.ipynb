```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
plt.title("global&shared")
g2s = np.loadtxt("global2shared.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('blockdim').mean().reset_index()
plt.plot(df2['blockdim'],df2['bandwidth'],label = "g2s")

g2s = np.loadtxt("shared2global.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('blockdim').mean().reset_index()
plt.plot(df2['blockdim'],df2['bandwidth'],label = "s2g")
plt.legend()
plt.xlabel("blockdim")
plt.ylabel("bandwidth (GB/s)")
plt.savefig("plots/global&sharedthreads.png")
plt.show()

plt.title("global&shared")
g2s = np.loadtxt("global2sharedblocks.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
plt.plot(df['griddim'],df['bandwidth'],label = "g2s")

g2s = np.loadtxt("shared2globalblocks.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
plt.plot(df['griddim'],df['bandwidth'],label = "s2g")
plt.legend()
plt.xlabel("griddim")
plt.ylabel("bandwidth (GB/s)")
plt.savefig("plots/global&sharedblocks.png")
plt.show()

plt.title("shared&register")
g2s = np.loadtxt("shared2register.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('blockdim').mean().reset_index()
plt.plot(df2['blockdim'],df2['bandwidth'],label = "s2r")

g2s = np.loadtxt("register2shared.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('blockdim').mean().reset_index()
plt.plot(df2['blockdim'],df2['bandwidth'],label = "r2s")
plt.legend()
plt.xlabel("blockdim")
plt.ylabel("bandwidth (GB/s)")
plt.savefig("plots/register&sharedthreads.png")

plt.show()

plt.title("shared&register")
g2s = np.loadtxt("shared2register.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('griddim').mean().reset_index()
plt.plot(df2['griddim'],df2['bandwidth'],label = "s2r")

g2s = np.loadtxt("register2shared.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","bandwidth"])
df2 = df.groupby('griddim').mean().reset_index()
plt.plot(df2['griddim'],df2['bandwidth'],label = "r2s")
plt.legend()
plt.xlabel("griddim")
plt.ylabel("bandwidth (GB/s)")
plt.savefig("plots/global&sharedblocks.png")

plt.show()

plt.title("shared2register conflict")
g2s = np.loadtxt("shared2register_conflict.txt", delimiter = " ")
df = pd.DataFrame(g2s,columns = ["griddim","blockdim","datasize","stride","clocks"])
plt.plot(df['stride'],df['clocks'],label = "s2g")
plt.xlabel("stride")
plt.ylabel("clocks")
plt.savefig("plots/shared2register_conflict.png")

plt.show()
```
