```python
import numpy as np
import matplotlib.pyplot as plt

Synctime = np.loadtxt("SyncTime.txt", delimiter = " ")
Asynctime = np.loadtxt("AsyncTime.txt", delimiter = " ")

STBlocks = np.array([a for a in Synctime if a[1] == 1])
STThreads = np.array([a for a in Synctime if a[0] == 1])
ASTBlocks = np.array([a for a in Asynctime if a[1] == 1])
ASTThreads = np.array([a for a in Asynctime if a[0] == 1])


plt.title("startuptime 1threadperblock")
plt.plot(STBlocks[:,0],STBlocks[:,2], label = "SyncTime")
plt.plot(ASTBlocks[:,0],ASTBlocks[:,2],label = "Asynctime")
plt.legend()
plt.xlabel("blocks")
plt.ylabel("startuptime")
plt.savefig("plots/Ex1TpBlock.png")
plt.show()
plt.title("startuptime 1block")
plt.plot(STThreads[:,1],STThreads[:,2],label = "SyncTime")
plt.plot(ASTThreads[:,1],ASTThreads[:,2], label = "Asynctime")
plt.legend()
plt.xlabel("threadsperblock")
plt.ylabel("startuptime")
plt.savefig("plots/Ex1Blocks.png")
plt.show()
```

```python
BreakEven = np.loadtxt("BreakEvenAsync.txt", delimiter = " ")

plt.plot(BreakEven[:,0],BreakEven[:,1])
plt.title("Asynctime for waiting Kernel")
plt.xlabel("WaitTime")
plt.ylabel("startuptime")
plt.savefig("plots/Ex2Async.png")
plt.show()

BreakEven = np.loadtxt("BreakEvenSync.txt", delimiter = " ")

plt.plot(BreakEven[:,0],BreakEven[:,1])
plt.title("Synctime for waiting Kernel")
plt.xlabel("WaitTime")
plt.ylabel("startuptime")
plt.savefig("plots/Ex2Sync-png")
plt.show()

```

```python
MemoryMallocH2D = np.loadtxt("MemoryMallocH2D.txt", delimiter = " ")
MemoryMallocD2H = np.loadtxt("MemoryMallocD2H.txt", delimiter = " ")
MemoryMallocHostH2D = np.loadtxt("MemoryMallocHostH2D.txt", delimiter = " ")
MemoryMallocHostD2H = np.loadtxt("MemoryMallocHostD2H.txt", delimiter = " ")

plt.plot(MemoryMallocH2D[:,0],MemoryMallocH2D[:,1],label = "Malloc H2D")
plt.plot(MemoryMallocD2H[:,0],MemoryMallocD2H[:,1],label = "Malloc D2H")
plt.plot(MemoryMallocHostH2D[:,0],MemoryMallocHostH2D[:,1],label = "Pinned H2D")
plt.plot(MemoryMallocHostD2H[:,0],MemoryMallocHostD2H[:,1], label = "Pinned D2H")
plt.title("Avg. Transfertime for 1000 DataTransfers")
plt.xlabel("Data (Byte)")
plt.ylabel("Time (us)")
plt.savefig("plots/MemorySpeed.png")
plt.legend()
plt.show()


plt.plot(MemoryMallocD2H[:,0],MemoryMallocD2H[:,1],label = "Malloc D2H")
plt.plot(MemoryMallocHostD2H[:,0],MemoryMallocHostD2H[:,1], label = "Pinned D2H")
plt.title("Avg. Transfertime for 1000 DataTransfers")
plt.xlabel("Data (Byte)")
plt.ylabel("Time (us)")
plt.savefig("plots/MemorySpeedD2H.png")
plt.legend()
plt.show()

plt.plot(MemoryMallocH2D[:,0],MemoryMallocH2D[:,1],label = "Malloc H2D")
plt.plot(MemoryMallocHostH2D[:,0],MemoryMallocHostH2D[:,1], label = "Pinned H2D")
plt.title("Avg. Transfertime for 1000 DataTransfers")
plt.xlabel("Data (Byte)")
plt.ylabel("Time (us)")
plt.savefig("plots/MemorySpeedH2D.png")
plt.legend()
plt.show()

```
