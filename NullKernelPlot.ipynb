```python
import numpy as np
import matplotlib.pyplot as plt

Synctime = np.loadtxt("SyncTime.txt", delimiter = " ")
Asynctime = np.loadtxt("AsyncTime.txt", delimiter = " ")

STBlocks = np.array([a for a in Synctime if a[1] == 1])
STThreads = np.array([a for a in Synctime if a[0] == 1])
ASTBlocks = np.array([a for a in Asynctime if a[1] == 1])
ASTThreads = np.array([a for a in Asynctime if a[0] == 1])


plt.title("startuptime 1threadperblock")
plt.plot(STBlocks[:,0],STBlocks[:,2], label = "SyncTime")
plt.plot(ASTBlocks[:,0],ASTBlocks[:,2],label = "Asynctime")
plt.legend()
plt.xlabel("blocks")
plt.ylabel("startuptime")
plt.show()
plt.title("startuptime 1block")
plt.plot(STThreads[:,1],STThreads[:,2],label = "SyncTime")
plt.plot(ASTThreads[:,1],ASTThreads[:,2], label = "Asynctime")
plt.legend()
plt.xlabel("threadsperblock")
plt.ylabel("startuptime")
plt.show()
```
